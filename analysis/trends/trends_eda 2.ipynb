```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(14.7,5.27)})
```

```python
file = pd.read_csv('../../data_collection/trends/data/monthly_interest_over_time.csv')
```

```python
file.columns
```

```python
example = file[['Anonymice_date', 'Anonymice', 'Anonymice NFT_date']].copy()
```

```python
example = example.set_index('Anonymice_date')
```

```python
example
```

```python
plt.figure()

sns.lineplot(data=example)
plt.xticks(rotation=90)
plt.xlabel("week_date")
plt.ylabel("interest")
plt.title('Anonymice interest over time')
plt.tight_layout()
plt.savefig('line_plots/demo/anonymice_solo.png')
```
