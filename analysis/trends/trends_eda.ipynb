```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(14.7,5.27)})
```

```python
file = pd.read_csv('../../data_collection/trends/data/daily_interest.csv')
```

```python
file.columns
```

```python
example = file[['Anonymice_date', 'Anonymice', 'Anonymice NFT_date', 'Anonymice NFT']].copy()
```

```python
example = example.set_index('Anonymice_date')
```

```python
example['Anonymice'] = example['Anonymice'] + example['Anonymice NFT']
example = example.drop(columns=['Anonymice NFT', 'Anonymice NFT_date'])
```

```python
example = example.reset_index()
```

```python
example['Anonymice_date'] = example['Anonymice_date'].apply(lambda x: pd.to_datetime(x))
# example.set_index('Anonymice_date')
```

```python
example
```

```python
plt.figure()

sns.lineplot(x='Anonymice_date', y='Anonymice', data=example)
plt.xticks(rotation=45)
plt.xlabel("date")
plt.ylabel("interest")
plt.title('Anonymice interest over time')
plt.tight_layout()
plt.savefig('line_plots/demo/anonymice_solo.png')
```
