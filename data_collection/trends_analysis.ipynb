### Separating _date columns

```python
date_col_l = []
for col in data.columns:
    if '_date' in col:
        date_col_l.append(col)
```

```python
date_df = data[date_col_l].copy()
for col_name in date_df.columns:
    data = data.drop(columns=col_name)
```
